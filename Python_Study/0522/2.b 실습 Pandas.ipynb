리뷰점수가 95점 이상이고, Australia와 New Zealand 에서 만들어진 와인에 대한 데이터프레임을 top_oceania_wines 이라는 이름의 변수로 저장.


```python
reviews
```

```python
reviews['points']>=95
```

```python
reviews['country']=='Australia'
```

```python
reviews['country']=='New Zealand'
```

```python
(reviews['country']=='Australia') | (reviews['country']=='New Zealand') # or은 조건문에서 사용
```

```python
reviews.loc[(reviews['points']>=95)&((reviews['country']=='Australia') | (reviews['country']=='New Zealand')),]
```

```python
reviews['country'].isin(['Australia','New Zealand'])
```

```python
reviews.loc[(reviews['points']>=95)&(reviews['country'].isin(['Australia','New Zealand'])),]
```

```python
top_oceania_wines=reviews[(reviews['points']>=95) & ((reviews['country']=='Australia')|(reviews['country']=='New Zealand'))]
top_oceania_wines
```
