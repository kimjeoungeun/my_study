- apply
  - 맵보다 적용할 수 있는 범위가 큰 것

```python
import pandas as pd
s=pd.Series([20,21,12],index=['London','New York','Helsinki'])
s
```

```python
def sub_custom_value(x,val):
    return x-val
```

```python
s.apply(sub_custom_value,args=(10,)) # 뒷칸 비워줘야 공간이 생김 가변공간
```

```python
def add_custom_values(x,**kwargs):
    for month in kwargs:
        x +=kwargs[month]
    return x
```

```python
s.apply(add_custom_values, june=30, july=20,august=25)
# 잘못된 수식 month라는 데이터가 없어서 다른 수치에 값을 더함
```

```python
frame.apply(lambda x : x.max()-x.min(), axis=1)
```

- Merge
    - key를 이용해 데이터의 row를 기준으로 연결시켜 합침(SQL의 join과 유사)

```python
data1=pd.DataFrame({'id':['01','02','03','04','05','06'],
                   'col1':np.random.randint(0,50,6),
                   'col2':np.random.randint(1000,2000,6)})
data1
```

```python
data2=pd.DataFrame({'id':['04','05','06','07'],
                    'col1':np.random.randint(1000,5000,4)})
data2
```

```python
# inner join(교집합)
pd.merge(data1, data2, on='id') # on은 기준
# col1이 겹치므로 x와 y로 표현
```

```python
# how: 어떤 방식으로 병합할 것인가?(inner, outer, left, right)
pd.merge(data1, data2, on='id', how='left')
```

```python
# outer join(합집합)
pd.merge(data1, data2, on='id', how='outer')
```

- cut

```python
ages=[20,35,67,54,54,1,2,8,4,9,15,20,75,74,14,52,36,25,34,35]
bins=[0,20,40,60,100]
```

```python
cuts=pd.cut(ages,bins) # 구간이 잘림
cuts
```

```python
cuts.categories # cut 메소드의 결과는 Categorical이라는 특수 객체에 존재
```

```python
cuts.codes
```

```python
group_names=['10대 이하', '20-30대', '40-50대', '60대 이상']
group=pd.cut(ages,bins, labels=group_names)
group
```

```python
# 구간을 균등한 길이로 나눔
pd.cut(ages,4,precision=3).value_counts()
# precision : 소수점 자릿수, 3이면 소수점 세번쨰에서 올림
```

```python
# 구간을 균등한 길이로 나눔
pd.qcut(ages,4).value_counts()
```

- 어떤 제작사가 제작한 영화의 총수익이 가장 높은지 알아보자.

- 원본 csv에서 달러 기호와 쉼표 기호를 유지하기 위해 열의 값을 텍스트로 가져옴.
- 열의 값을 십진수로 변환하려면 두 가지 기호('$',',')를 모두 제거하고 빈 텍스트로 바꿔야함

```python
df['Gross']=df['Gross'].str.replace("$","",regex=False).str.replace(",","",regex=False).astype(float)
```

```python
df['Gross'].mean()
```
